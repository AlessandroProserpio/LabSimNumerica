```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
```

## Exercise 11.1
We try to generate a simple neural network of one neuron able to fit a linear function $f(x):=mx+q$ changning the number of epochs $N_{{\rm epochs}}\in\{20,50, 100\}$, the size of the training set $N_{{\rm train}}\in\{150, 300\}$ and the variance of the gaussian noise $\sigma\in\{0, 0.5, 1\}$.

The loss function is SME and the optimizing algorithm is SGD.

```python
m=2 
b=1 
np.random.seed(0)
x_valid = np.random.uniform(-1, 1, 100)
x_valid.sort()
y_valid = np.random.normal(m * x_valid + b, sigma)
y_target = m * x_valid + b 

epochs=[20, 50, 100]
trains=[150, 300]
sigmas=[0, 0.5, 1.]

i=0
fig, c=plt.subplots(nrows=len(epochs)*len(trains)*len(sigmas), ncols=2)

for sigma in sigmas:
    for epoch in epochs:
        for train in trains:
            x_train = np.random.uniform(-1, 1, train)
            y_train = np.random.normal(m * x_train + b, sigma)
            model = tf.keras.Sequential()
            model.add(Dense(1, input_shape=(1,)))
            model.compile(optimizer='sgd', loss='mse', metrics=['mse'])
            history = model.fit(x=x_train, y=y_train, batch_size=32, epochs=epoch, shuffle=True, validation_data=(x_valid, y_valid))
            y_predicted = model.predict(x_valid)
            c[i, 0].scatter(x_valid, y_predicted, label="Predicted")
            c[i, 0].plot(x_valid, y_target, color="red", label="Target")
            c[i, 0].legend()
            c[i, 0].set_ylabel('f(x)')
            c[i, 0].set_xlabel('x')
            c[i, 0].grid(True)
            c[i, 0].set_title(f"Target & predicted data \n N epochs = {epoch} ; N train = {train} ; $\sigma$= {sigma}")
            c[i, 1].plot(history.history['loss'])
            c[i, 1].plot(history.history['val_loss'])
            c[i, 1].set_title('Model loss')
            c[i, 1].set_ylabel('Loss')
            c[i, 1].set_xlabel('Epoch')
            c[i, 1].legend(['Train', 'Test'], loc='best')
            c[i, 1].grid(True)
            i+=1
plt.subplots_adjust(left=0, bottom=-15, right=2, top=1.2, wspace=0.2, hspace=0.4)
```

Notice (scroll down) that at different $\sigma$-s, the error curves show a similar behaviour as the other parameters change: they start off almost parallel, then - as the number of epochs increases - the training error stabilizes around a value $\sigma$-related while the test error settles about the value $b=1$ which is the $y$-intercept (which accordingly enters in the output of the network as a bias). $50$ epochs are quite enough to get the errors to stabilize (for the bigger $\sigma$-s the training set should be the bigger ones).

## Exercise 11.2
We try to build a neural network able to fit the polynomial: $$ P(x):=3x^3-2x^2-3x+4$$ We add some gaussian noise to the training and validation data $\{x_i, y_i\}_{i=1}^N$, respetively of size $N_T=1300$ and $N_V=150$. The $x_i$-s lie in the interval $[-1,1]$.

We will build different networks and compare their results by confronting the predicted curves for points both in $[-1,1]$ and outside as well as the loss functions for the training and validation sets. In particular, we will create $24$ networks for four choices of the number of layers ($3,5,10,15$), three choices for the number of neurons in each layer ($4,8,12$, except of course for the input and output layers) and two activation functions (<code>selu</code> from the open bound functions, <code>tanh</code> from the ones with plateau). We will fix the number of epochs to $20$, the cost function to MSE and the optimizer to SGD.

```python
layers=[3, 5, 10, 15]
neurons=[4, 8, 12]
active=["selu", "tanh"]

sigma=0.6

x_train=np.random.uniform(-1, 1, 1300)
x_valid=np.random.uniform(-1, 1, 150)
x_valid.sort()
y_train=np.random.normal( 3*x_train**3-2*x_train**2-3*x_train+4, sigma)
y_valid=np.random.normal( 3*x_valid**3-2*x_valid**2-3*x_valid+4, sigma)
```

```python
nepochs=20
xpred=np.arange(-1.5, 1.5, 0.1)
ytarget=3*xpred**3-2*xpred**2-3*xpred+4

i=0
fig, c=plt.subplots(nrows=len(layers)*len(neurons)*len(active), ncols=2)
for layer in layers:
    for neuron in neurons:
        for a in active:
            model=tf.keras.Sequential()
            model.add(Dense(neuron, input_shape=(1,), activation=a))
            for k in range(layer-2):
                model.add(Dense(neuron, activation=a))
            model.add(Dense(1))
            model.compile(optimizer='sgd', loss='mse', metrics=['mse'])
            history=model.fit(x=x_train, y=y_train, batch_size=32, epochs=nepochs, shuffle=True, validation_data=(x_valid, y_valid))
            y_predicted = model.predict(xpred)
            c[i, 0].scatter(xpred, y_predicted, label="Predicted")
            c[i, 0].plot(xpred, ytarget, color="red", label="Target")
            c[i, 0].legend()
            c[i, 0].grid(True)
            c[i, 0].set_title(f"Target & predicted data \n Layers = {layer} ; Neurons = {neuron} ; Activation= {a}")
            c[i, 1].plot(history.history['loss'])
            c[i, 1].plot(history.history['val_loss'])
            c[i, 1].set_title('Model loss')
            c[i, 1].set_ylabel('Loss')
            c[i, 1].set_xlabel('Epoch')
            c[i, 1].legend(['Train', 'Test'], loc='best')
            c[i, 1].grid(True)
            i+=1
plt.subplots_adjust(left=0, bottom=-20, right=2, top=1.2, wspace=0.2, hspace=0.4)
```

As for the curves (scroll down), with few exceptions, we see that the predicted values are reasonably close to the target for datas inside the training interval, while all the networks dramatically fail outside $[-1,1]$. In general, the <code>selu</code> networks behave better than the <code>tanh</code> ones. The former perform better predictions especially for $x<0$.

As for the errors, we see that three layer seem to be enough to get a well-trained neural network for the <code>selu</code> activation function: the errors on training and testing sets are close to each other already in such cases while, as we raise the number of layers, we see big oscillations in the test error as compared to the training. This shows that the previous model where enough complex and the increase in the number of parameters starts causing overfitting. On the contrary, the <code>tanh</code> networks show remarkable oscillations in the testing error only for the biggest number of layers we have chosen (i.e. it takes more layer to reach the overfitting region for such networks). Notice, however, that in many cases, for the same number of layers and neurons, the <code>tanh</code> networks display slightly higher values for the cost functions as compared to those of the <code>selu</code> networks.

## Exercise 11.3
Using the results of the previous exercises, we build a neural network able to fit the surface defined by:
$$ f(x,y):=\sin(x^2+y^2) $$ for $(x,y)\in\bigl[-\frac{3}{2},\frac{3}{2}\bigr]^2$.

We will once again try $12$ different networks varying the number of layers ($5,10,15,20$) and the number of neurons per layer ($5, 10,15$) for fixed training and testing sample sizes (respectively $N_T=1500$ and $N_V=200$, both with a gaussian noise of variance $\sigma=0.3$), activation function (<code>selu</code>) and number of epochs ($50$). We plot training and validation errors as a function of the number of epochs and compare the target to the prediction for a sample of equally-spaced points on the half-diagonal $\{x\in\bigl[0,\frac{3}{2}\bigr],\,y=x\}$ as a function of the distance to the origin $r:=\sqrt{x^2+y^2}$.

```python
sigma=0.3
x_train=np.random.uniform(-1.5, 1.5, 1500)
y_train=np.random.uniform(-1.5, 1.5, 1500)
x_valid=np.random.uniform(-1.5, 1.5, 200)
y_valid=np.random.uniform(-1.5, 1.5, 200)
x_valid.sort()
y_valid.sort()
xpred=np.arange(0, 1.5, 0.05)


f_target=np.sin(2*xpred**2)
f_train=np.random.normal(np.sin(x_train**2+y_train**2), sigma)
f_valid=np.random.normal(np.sin(x_valid**2+y_valid**2), sigma)
```

```python
layers=[5, 10, 15, 20]
neurons=[5, 10, 15]

i=0
fig, c=plt.subplots(nrows=len(layers)*len(neurons), ncols=2)
for layer in layers:
    for neuron in neurons:
        model=tf.keras.Sequential()
        model.add(Dense(neuron, input_shape=(2,), activation='selu'))
        for k in range(layer-2):
            model.add(Dense(neuron, activation='selu'))
        model.add(Dense(1))
        model.compile(optimizer='sgd', loss='mse', metrics=['mse'])
        history=model.fit(x=np.array([x_train, y_train]).T, y=f_train, batch_size=32, epochs=50, shuffle=True, 
                          validation_data=(np.array([x_valid, y_valid]).T, f_valid))
        f_predicted = model.predict(np.array([xpred, xpred]).T)
        c[i, 0].scatter(np.sqrt(2*xpred**2), f_predicted, color='r', label="Predicted")
        c[i, 0].plot(np.sqrt(2*xpred**2), f_target, label="Target")
        c[i, 0].grid(True)
        c[i, 0].legend()
        c[i, 0].set_ylabel('f ( r )')
        c[i, 0].set_xlabel('r')
        c[i, 0].set_title(f"Target curve & predicted data \n Layers = {layer} ; Neurons = {neuron}")
        c[i, 1].plot(history.history['loss'])
        c[i, 1].plot(history.history['val_loss'])
        c[i, 1].set_title('Model loss')
        c[i, 1].set_ylabel('Loss')
        c[i, 1].set_xlabel('Epoch')
        c[i, 1].legend(['Train', 'Test'], loc='best')
        c[i, 1].grid(True)
        i+=1
plt.subplots_adjust(left=0, bottom=-10, right=2, top=1.2, wspace=0.2, hspace=0.4)
```

The network with $10$ layer and $15$ neurons per layer yields the best results both in the errors and in the comparison between target and predicted points. We plot a 3D graph for such NN:

```python
from mpl_toolkits import mplot3d

model=tf.keras.Sequential()
model.add(Dense(neuron, input_shape=(2,), activation='selu'))
for k in range(15-2):
        model.add(Dense(neuron, activation='selu'))
model.add(Dense(1))
model.compile(optimizer='sgd', loss='mse', metrics=['mse'])
history=model.fit(x=np.array([x_train, y_train]).T, y=f_train, batch_size=32, epochs=50, shuffle=True, validation_data=(np.array([x_valid, y_valid]).T, f_valid))
x_predicted=np.random.uniform(-1.5, 1.5, 1000)
y_predicted=np.random.uniform(-1.5, 1.5, 1000)
f_predicted=model.predict(np.array([x_predicted, y_predicted]).T)

x=np.outer(np.linspace(-1.5,1.5,30), np.ones(30))
y=x.copy().T
z=np.sin(x**2+y**2)
fig = plt.figure()
fig=plt.figure(figsize=(10, 8))

ax = plt.axes(projection='3d')
ax.plot_surface(x, y, z, cmap='viridis', alpha=0.5)
ax.scatter(x_predicted, y_predicted, f_predicted, color='r')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f ( x , y )')
ax.set_title("Target surface & predicted data")
ax.view_init(45,45)
plt.show()
```
